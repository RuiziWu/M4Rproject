```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
pDist_H1_05_bd1 = np.loadtxt("pDist_H1_05_bd1.csv", dtype = float, delimiter = ",")
```

```python
fig = plt.figure(figsize =(8, 6))
# fig.suptitle("H1, samplesize=100, bandwidth=1, mean=0.5", fontsize = 16)

ax = fig.add_subplot(111)
ax.boxplot(pDist_H1_05_bd1.T)
ax.set_xticklabels([1, 10, 50, 100, 500, 1000, 1500])
ax.set_xlabel("dimension")
ax.set_ylabel("p value")

plt.show()
```
