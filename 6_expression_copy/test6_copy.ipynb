```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ksd_value_10 = np.loadtxt("KSD_value_10.csv", delimiter = ",", dtype = float)
ksd_variance_10 = np.loadtxt("KSD_variance_10.csv", delimiter = ",", dtype = float)
ksd_value_25 = np.loadtxt("KSD_value_25.csv", delimiter = ",", dtype = float)
ksd_variance_25 = np.loadtxt("KSD_variance_25.csv", delimiter = ",", dtype = float)
```

```python
ksd_cheb_10 = np.loadtxt("KSD_cheb_10.csv", delimiter = ",", dtype = float)
ksd_prob_10 = np.loadtxt("KSD_prob_10.csv", delimiter = ",", dtype = float)
ksd_cheb_25 = np.loadtxt("KSD_cheb_25.csv", delimiter = ",", dtype = float)
ksd_prob_25 = np.loadtxt("KSD_prob_25.csv", delimiter = ",", dtype = float)
```

```python
ksd_cheb = np.append(ksd_cheb_10, ksd_cheb_25)
ksd_prob = np.append(ksd_prob_10, ksd_prob_25)
```

```python
x_axis = np.array(["1", "5", "10", "15", "20", "25"])

fig = plt.figure(figsize=(8, 6))

ax = fig.add_subplot(111)
ax.plot(x_axis, ksd_prob, 'b', label="P(KSD > 0.05)")
ax.plot(x_axis, ksd_cheb, 'g', label="chebshev bound")
ax.legend(bbox_to_anchor=(1, 0.5), fontsize=12, title_fontsize=12)

plt.show()
```

```python

ksd_prob_10c = 1 - ksd_prob_10
ksd_cheb_10c = 1 - ksd_prob_10
ax = fig.add_subplot(122)
ax.plot(x_axis, ksd_prob_10c, 'b', label="P(KSD <= 0.05)")
ax.plot(x_axis, ksd_cheb_10c, 'g', label="1 - chebshev bound")
ax.legend(bbox_to_anchor=(1, 0.5), fontsize=12, title_fontsize=12)
```
