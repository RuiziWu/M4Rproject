```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
P_ksd_test = np.loadtxt("P_ksd_test.csv", dtype = float, delimiter = ",")
E_ksd_test = np.loadtxt("True_expectation_test.csv", dtype = float, delimiter = ",")
V_ksd_test = np.loadtxt("True_variance_test.csv", dtype = float, delimiter = ",")
```

```python
U_bound = V_ksd_test / (0.001**2)
```

```python
x_axis = np.array(["1", "10", "50", "100", "500", "1000"])

fig = plt.figure(figsize=(16, 6))

ax1 = fig.add_subplot(121)
ax1.plot(x_axis, P_ksd_test, 'b', label="P(KSD > KSD* + 0.001)")
ax1.plot(x_axis, U_bound, 'g', label="chebshev bound")
ax1.legend(bbox_to_anchor=(1, 0.5), fontsize=12, title_fontsize=12)

ax2 = fig.add_subplot(122)
ax2.plot(x_axis, P_ksd_test, 'b', label="P(KSD > KSD* + 0.001)")
ax1.legend(bbox_to_anchor=(1, 0.5), fontsize=12, title_fontsize=12)

plt.show()
```
